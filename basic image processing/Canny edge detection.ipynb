```
image=cv2.imread('/content/Sunflower.jpg')
image=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.imshow(image)
```

```
image_copy=np.copy(image)
gray_image=cv2.cvtColor(image_copy,cv2.COLOR_RGB2GRAY)
plt.imshow(gray_image,cmap='gray')
```

#Canny Edge Detection

```
lower=120
high=240
edge=cv2.Canny(gray_image,lower,high)
plt.imshow(edge,cmap='gray')
```

```
wide=cv2.Canny(gray_image,30,100)
tight=cv2.Canny(gray_image,180,240)
f,(ax1,ax2)=plt.subplots(1,2,figsize=(20,20))
ax1.set_title('wide')
ax1.imshow(wide,cmap='gray')
ax2.set_title('tight')
ax2.imshow(tight,cmap='gray')
```

###in this scenario wide outperform tight in detecting edges correctly
