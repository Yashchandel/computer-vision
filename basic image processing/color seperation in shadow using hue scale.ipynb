```
import numpy as np
import matplotlib.pyplot as plt
import cv2
%matplotlib inline
```

```
image=cv2.imread('/content/WhatsApp Image 2020-11-07 at 6.45.20 PM.jpeg')
image_copy=np.copy(image)
image_copy=cv2.cvtColor(image_copy,cv2.COLOR_BGR2RGB)
plt.imshow(image_copy)
```

##Plot Color Channel

```
##Plot color channel
r=image_copy[:,:,0]
g=image_copy[:,:,1]
b=image_copy[:,:,2]
f,(ax1,ax2,ax3)=plt.subplots(1,3,figsize=(20,10))
ax1.set_title('Red')
ax1.imshow(r,cmap='gray')
ax2.set_title('Green')
ax2.imshow(g,cmap='gray')
ax3.set_title('blue')
ax3.imshow(b,cmap='gray')
```

```
#Convert RGB to HSV
hsv=cv2.cvtColor(image_copy,cv2.COLOR_RGB2HSV)
h=hsv[:,:,0]
s=hsv[:,:,1]
v=hsv[:,:,2]
f,(ax1,ax2,ax3)=plt.subplots(1,3,figsize=(20,10))
ax1.set_title('Hue')
ax1.imshow(h,cmap='gray')
ax2.set_title('Saturation')
ax2.imshow(s,cmap='gray')
ax3.set_title('Value')
ax3.imshow(v,cmap='gray')
```

##Define pink and hue selection

```
#boundaries for pink
lower_pink=np.array([180,0,100])
higher_pink=np.array([255,255,230])
```

```
#boundaries for hue
lower_hue=np.array([160,0,0])
higher_hue=np.array([180,255,255])
```

```
#mask the image
mask_rgb=cv2.inRange(image_copy,lower_pink,higher_pink)
masked_image=np.copy(image)
masked_image[mask_rgb==0]=[0,0,0]
plt.imshow(masked_image)
```

```
mask_hsv=cv2.inRange(hsv,lower_hue,higher_hue)
masked=np.copy(image)
masked[mask_hsv==0]=[0,0,0]
plt.imshow(masked)
```
