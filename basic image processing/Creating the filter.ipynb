```
image=cv2.imread('/content/San_Francisco_City_Hall_2.jpg')
image=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.imshow(image)
```

###CONVERT image to gray scale

```
image_copy=np.copy(image)
gray_image=cv2.cvtColor(image_copy,cv2.COLOR_RGB2GRAY)
plt.imshow(gray_image,cmap='gray')
```

###Create Custom kernel

```
#detecting vertical edges
#common naed as sobel filter
sobel_x=np.array([[-1,0,1],[-2,0,2],[-1,0,1]])
filter_image=cv2.filter2D(gray_image,-1,sobel_x)#-1 indicate same dim of output as input
plt.imshow(filter_image,cmap='gray')
```

```
import matplotlib
matplotlib.image.imsave('/content/San_Francisco_City_Hall.jpg',filter_image,cmap='gray')
```

###Create binary image


```
retval,binary_image=cv2.threshold(filter_image,100,255,cv2.THRESH_BINARY)
plt.imshow(binary_image,cmap='gray')
```

```
matplotlib.image.imsave('/content/San_Francisco_City_Hall_1.jpg',binary_image,cmap='gray')
```
