```
import numpy as np
import matplotlib.pyplot as plt
import cv2
%matplotlib inline
```

```
image=cv2.imread('/content/Monarch_In_May.jpg (3008Ã—2000) - Google Chrome 17-11-2020 15_56_29.png')
image=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.imshow(image)
```

##Reshaping image for k mean clustering

```
image_copy=np.copy(image)
pixel_val=image_copy.reshape(-1,3)
pixel_val=np.float32(pixel_val)
```

#Implementing K means

```
criteria=(cv2.TERM_CRITERIA_EPS+cv2.TERM_CRITERIA_MAX_ITER,10,1.0)
k=2
retval,labels,centres=cv2.kmeans(pixel_val,k,None,criteria,10,cv2.KMEANS_RANDOM_CENTERS)
#converting data to 8 bit unit
centres=np.uint8(centres)
segmented_data=centres[labels.flatten()]
# reshaping data inti original image dimension
segmented_image=segmented_data.reshape((image_copy.shape))
labels_reshape=labels.reshape(image_copy.shape[0],image_copy.shape[1])
plt.imshow(segmented_image)
```

```
plt.imshow(labels_reshape==1,cmap='gray')
```

```
plt.imshow(labels_reshape==0,cmap='gray')
```

```
masked_image=np.copy(image_copy)
masked_image[labels_reshape==0]=[0,0,0]
plt.imshow(masked_image)
# you can perform many operation on it
```
